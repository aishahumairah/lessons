### Discussion 2: How to Learn Independently

**Opening question:** Think of something technical you taught yourself (does not have to be coding). How did you do it? What worked?

---

**Discussion prompts:**

1. When you get completely stuck on a technical problem, what is your first instinct? What *should* it be?

2. There is a difference between copying a working answer and understanding why it works. When is copying fine? When is it harmful to your learning?

3. This course covers 12 weeks of material. How will you handle a week where life gets in the way and you fall behind?

4. What is the difference between *I finished the exercise* and *I understand this well enough to use it on a new problem*?

---

**Habits that experienced engineers swear by** (for discussion, not a lecture):

- Read error messages fully before Googling
- Rubber duck debugging — explain the problem out loud before asking for help
- Spend 20 minutes stuck before reaching out; spend no more than 45 minutes before reaching out
- Done means you can explain it without looking at your notes

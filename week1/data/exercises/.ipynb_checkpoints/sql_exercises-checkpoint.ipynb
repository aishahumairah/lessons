**Exercise 2.9** — List Engineering employees who earn more than $100,000 OR were hired before 2020.

*Expected: 4 rows*

> **Hint — AND/OR precedence:** `WHERE a AND b OR c` is read as `WHERE (a AND b) OR c`.
> To get *Engineering employees* who satisfy either condition, you need:
> `WHERE department = 'Engineering' AND (salary > 100000 OR hire_date < '2020-01-01')`
> The parentheses are required — without them the query returns different rows.

## Solutions

Only check after attempting each exercise.

---

**1.1** `SELECT first_name, department, salary FROM employees;`

**1.2** `SELECT * FROM employees;`

**1.3** `SELECT first_name, salary AS annual_pay FROM employees;`

**1.4** `SELECT first_name, salary, ROUND(salary / 12.0, 2) AS monthly_salary FROM employees;`

---

**2.1** `SELECT * FROM employees WHERE department = 'HR';`

**2.2** `SELECT * FROM employees WHERE salary = 65000;`

**2.3** `SELECT * FROM employees WHERE salary >= 90000;`

**2.4** `SELECT * FROM employees WHERE NOT department = 'Engineering';`

**2.5** `SELECT * FROM employees WHERE department = 'Marketing' OR department = 'Sales';`

**2.6** `SELECT * FROM employees WHERE department IN ('Engineering', 'Marketing', 'Sales');`

**2.7** `SELECT * FROM employees WHERE salary BETWEEN 70000 AND 80000;`

**2.8** `SELECT * FROM employees WHERE first_name LIKE 'A%' OR first_name LIKE 'B%' OR first_name LIKE 'C%';`

**2.9**
```sql
SELECT * FROM employees
WHERE department = 'Engineering'
  AND (salary > 100000 OR hire_date < '2020-01-01');
```

**2.10** `SELECT * FROM employees WHERE hire_date IS NOT NULL;`

---

**3.1** `SELECT * FROM employees ORDER BY salary ASC;`

**3.2** `SELECT * FROM employees ORDER BY salary DESC;`

**3.3** `SELECT * FROM employees ORDER BY department ASC, salary DESC;`

**3.4** `SELECT * FROM employees ORDER BY hire_date ASC;`

---

**4.1** `SELECT * FROM employees LIMIT 5;`

**4.2** `SELECT * FROM employees ORDER BY salary DESC LIMIT 3;`

**4.3** `SELECT * FROM employees ORDER BY hire_date DESC LIMIT 3;`

---

**5.1** `SELECT DISTINCT department FROM employees ORDER BY department;`

**5.2** `SELECT DISTINCT city FROM employees ORDER BY city;`

**5.3** `SELECT DISTINCT department, city FROM employees ORDER BY department, city;`

**5.4** `SELECT COUNT(DISTINCT department) AS num_departments FROM employees;`

**5.5** `SELECT COUNT(DISTINCT city) AS num_cities FROM employees;`

---

**6.1**
```sql
SELECT first_name, last_name, salary, hire_date
FROM employees
WHERE department = 'Engineering'
ORDER BY salary DESC
LIMIT 3;
```

**6.2**
```sql
SELECT first_name, last_name, salary
FROM employees
WHERE city = 'Chicago' AND is_active = TRUE
ORDER BY salary DESC;
```

**6.3**
```sql
SELECT first_name, last_name, hire_date
FROM employees
WHERE hire_date BETWEEN '2020-01-01' AND '2021-12-31'
ORDER BY hire_date ASC;
```

**6.4**
```sql
SELECT DISTINCT department, city
FROM employees
ORDER BY department, city;
```

**6.5**
```sql
SELECT first_name, last_name, salary, hire_date
FROM employees
WHERE salary > 80000
  AND hire_date < '2021-01-01'
ORDER BY hire_date ASC;
```

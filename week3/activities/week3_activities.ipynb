## 1. Group Discussion

Choose one person to take notes. Spend 10–12 minutes per topic.

### Discussion Topic 1 — Python vs SQL: When to Use Which?

You now know both SQL and Python. They overlap — you can filter, group, and aggregate in either language.

**Opening question:** You have a table with 500,000 sales rows. You need to find the top 10 customers by total spend. Would you do this in SQL or Python? Why?

**Dig deeper:**
- SQL runs on the database server, where the data already lives. Python pulls the data to your machine first. When does that matter?
- When is Python the better choice even when SQL could do the job?
- Have you seen tools that blur this line? (Hint: pandas DataFrames can connect directly to databases.)

**Share:** Each group gives one scenario where SQL wins and one where Python wins.

### Discussion Topic 2 — What Makes a Cleaning Pipeline Trustworthy?

You built a pipeline that reads messy data and outputs a clean file. But how do you know it is correct?

**Opening question:** If you run the pipeline tomorrow and it produces 71 rows instead of 73, how would you find out why?

**Dig deeper:**
- What checks would you add to the pipeline to catch problems automatically?
- How do you test a cleaning function? What inputs would you try?
- What is the difference between a cleaning step that silently drops a row and one that logs a warning?
- Who should be told when the pipeline produces unexpected output?

**Share:** Each group names the one check they would add to their pipeline first.

### Discussion Topic 3 — Reusable Code vs One-Off Scripts

In this week you wrote cleaning functions that you called multiple times across multiple files. That is the beginning of a reusable library.

**Opening question:** When is it worth turning a one-off script into a reusable function?

**Dig deeper:**
- The `clean_price` function in this week will appear in almost every pipeline you ever write. What would happen if it was buried inside a script and not in a shared module?
- What makes a function easy for someone else to use without reading the code inside it?
- In professional data teams, cleaning logic is sometimes duplicated across ten different scripts, with small differences. What problems does that cause?

**Share:** Each group describes one piece of reusable logic from this week they would want to save and use again in week 4.

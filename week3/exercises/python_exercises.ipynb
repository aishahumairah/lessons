### Exercise 8
```python
def validate_order(row):
    if row['quantity'] <= 0:
        return False, f'Invalid quantity: {row["quantity"]}'
    if row['unit_price'] < 0:
        return False, f'Negative price: {row["unit_price"]}'
    expected = row['quantity'] * row['unit_price']
    if abs(row['total'] - expected) > 0.01:
        return False, f'Wrong total: {row["total"]} (expected {expected})'
    return True, 'OK'
```

def revenue_by_category(df):
    return (
        df.groupby('category')
        .agg(orders=('order_id', 'count'), revenue=('total', 'sum'))
        .sort_values('revenue', ascending=False)
        .round(2)
    )

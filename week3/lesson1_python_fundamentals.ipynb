## 1.1 Variables and Data Types

A **variable** stores a value. You name it, assign a value with `=`, and use it later.

Python has five core types you will use constantly:

| Type | Example | What it holds |
|------|---------|---------------|
| `int` | `42` | Whole numbers |
| `float` | `149.99` | Decimal numbers |
| `str` | `'Alice'` | Text |
| `bool` | `True` | Yes or No |
| `None` | `None` | Nothing / missing |

```python
order_id    = 1001
price       = 149.99
customer    = 'Alice Johnson'
is_paid     = True
refund_date = None

print(order_id)
print(price)
print(customer)
print(is_paid)
print(refund_date)
```

```python
# type() tells you what kind of value a variable holds
print(type(order_id))     # int
print(type(price))        # float
print(type(customer))     # str
print(type(is_paid))      # bool
print(type(refund_date))  # NoneType
```

### Converting between types

When Python reads a CSV file, **every value comes in as a string** — even numbers. You must convert them before you can do maths. `int()`, `float()`, and `str()` convert between types.

```python
# Everything from a CSV starts as a string
quantity_str = '3'      # this is text, not a number
price_str    = '149.99'

# You cannot do maths on strings
# print(quantity_str * price_str)  # TypeError — would crash

# Convert first
quantity = int(quantity_str)
price    = float(price_str)
total    = quantity * price
print(f'Total: ${total:.2f}')   # $449.97

# The danger: '10' > '9' gives the WRONG answer because strings sort alphabetically
print('10' > '9')    # False — string comparison ('1' comes before '9')
print(10 > 9)        # True  — number comparison

# This is why converting CSV columns to numbers before comparing is critical
```

## 1.2 String Operations

In data engineering, most of your time is spent fixing messy text: removing extra spaces, fixing casing, stripping currency symbols.

Python strings have built-in methods for all of this. A method is called with a dot: `value.method()`.

```python
# Real problem: categories in the CSV have inconsistent casing and whitespace
raw = '  ELECTRONICS  '

print(raw.strip())           # remove leading and trailing spaces
print(raw.strip().lower())   # lowercase
print(raw.strip().title())   # Title Case — the format we want

# You can chain methods: each one returns a string, and the next runs on that
cleaned = raw.strip().title()
print(cleaned)  # 'Electronics'
```

```python
# Real problem: price is stored as '$1,200.00' but we need the number 1200.0
price_str = '$1,200.00'

# Step by step
step1 = price_str.replace('$', '')   # '1,200.00'
step2 = step1.replace(',', '')        # '1200.00'
price = float(step2)                  # 1200.0
print(price)

# All in one line:
price = float(price_str.replace('$', '').replace(',', ''))
print(price)
```

```python
# .split() breaks a string into a list
# .join() combines a list back into a string

# Split a full name into first and last
full_name = 'Alice Johnson'
parts = full_name.split(' ')   # split on space
print(parts)                   # ['Alice', 'Johnson']
print(parts[0])                # 'Alice'
print(parts[-1])               # 'Johnson'

# Split a date string into components
date_str = '2023-07-14'
year, month, day = date_str.split('-')
print(f'Year: {year}, Month: {month}, Day: {day}')

# .join() is the reverse of .split()
words = ['Home', '&', 'Kitchen']
combined = ' '.join(words)
print(combined)   # 'Home & Kitchen'
```

```python
# .startswith() and .endswith() — check how a string begins or ends
filename = 'sales_q1.csv'

print(filename.endswith('.csv'))    # True — useful for filtering file lists
print(filename.startswith('sales')) # True
print(filename.startswith('q1'))    # False

# Practical use: process only CSV files, skip others
files = ['sales_q1.csv', 'notes.txt', 'sales_q2.csv', 'README.md']
csv_files = [f for f in files if f.endswith('.csv')]
print(csv_files)   # ['sales_q1.csv', 'sales_q2.csv']
```

```python
# f-strings: embed values inside text
product   = 'Laptop'
quantity  = 3
unit_price = 1200.00
total = quantity * unit_price

print(f'Product: {product}')
print(f'Quantity: {quantity}')
print(f'Total: ${total:,.2f}')   # $3,600.00 — , adds thousands separator, .2f = 2 decimal places
```

## 1.3 Lists

A **list** holds multiple values in order. In data engineering, a list often holds a collection of rows, column values, or file names to process.

```python
regions = ['North', 'South', 'East', 'West']

print(regions)        # the whole list
print(regions[0])     # first item (counting starts at 0)
print(regions[-1])    # last item
print(len(regions))   # 4

# Add an item
regions.append('Central')
print(regions)

# Loop over a list
for region in regions:
    print(f'  Region: {region}')
```

```python
# List comprehension: build a new list from an existing one
# Format: [expression for item in list]

raw_categories = ['ELECTRONICS', '  Clothing  ', 'HOME & KITCHEN']
cleaned = [c.strip().title() for c in raw_categories]
print(cleaned)   # ['Electronics', 'Clothing', 'Home & Kitchen']
```

```python
# Filtering comprehension: add an if condition to keep only matching items
# Format: [expression for item in list if condition]

quantities = ['2', '-1', '1', '0', '3', 'N/A']

# Keep only positive numeric quantities
valid = []
for q in quantities:
    try:
        if int(float(q)) > 0:
            valid.append(int(float(q)))
    except ValueError:
        pass  # skip non-numeric values

print(valid)   # [2, 1, 3]

# The same thing as a comprehension (more compact, same result)
valid_prices = [35.0, 0.0, 1200.0, -5.0, 150.0]
positives = [p for p in valid_prices if p > 0]
print(positives)   # [35.0, 1200.0, 150.0]
```

```python
# sorted() returns a sorted copy — the original list is unchanged
products = ['Wireless Headphones', 'Laptop', 'Gaming Mouse', 'Keyboard']
print(sorted(products))            # alphabetical A-Z
print(sorted(products, reverse=True))  # Z-A

# Sort by a computed value using key=
prices = [('Laptop', 1200), ('Mouse', 35), ('Keyboard', 75), ('Monitor', 400)]
by_price = sorted(prices, key=lambda item: item[1])
print(by_price)   # sorted cheapest to most expensive
```

## 1.4 Dictionaries

A **dictionary** maps keys to values. Each row of a CSV file becomes a dictionary when you read it with Python — the column names are the keys.

```python
# A single sales row as a dictionary
row = {
    'order_id':   1001,
    'customer':   'Alice Johnson',
    'product':    'Laptop',
    'quantity':   2,
    'unit_price': 1200.00
}

print(row['customer'])   # Alice Johnson
print(row['quantity'])   # 2

# Add a new key
row['total'] = row['quantity'] * row['unit_price']
print(row['total'])   # 2400.0

# Loop over all keys and values
for key, value in row.items():
    print(f'  {key}: {value}')
```

```python
# .get() is safer than [] when a key might be missing
print(row.get('region'))              # None — key absent, no crash
print(row.get('region', 'Unknown'))   # 'Unknown' — custom default
```

```python
# A dictionary as a lookup table — one of the most common patterns in pipelines
# Use it to correct typos or map codes to readable names

TYPO_MAP = {
    'electronics':    'Electronics',
    'electrnics':     'Electronics',   # typo from Q2 data
    'clothing':       'Clothing',
    'home & kitchen': 'Home & Kitchen',
    'home & kithen':  'Home & Kitchen', # typo from Q2 data
}

def correct_category(raw):
    key = raw.strip().lower()          # normalise before lookup
    return TYPO_MAP.get(key, raw.strip().title())  # fall back to title case if not in map

test_values = ['ELECTRONICS', 'Electrnics', '  clothing  ', 'Home & Kithen', 'Unknown Cat']
for v in test_values:
    print(f'  {v!r:25} -> {correct_category(v)}')
```

## 1.5 Control Flow

`if` / `elif` / `else` lets your code make decisions. `for` loops repeat an action for each item in a sequence.

```python
# Classify a row's total revenue into tiers
total = 1200.00

if total >= 500:
    tier = 'High'
elif total >= 100:
    tier = 'Medium'
else:
    tier = 'Low'

print(f'${total:.2f} is a {tier} order')
```

```python
# Loop with enumerate() to get the index alongside the value
files = ['sales_q1.csv', 'sales_q2.csv', 'sales_q3.csv']

for i, filename in enumerate(files, start=1):
    print(f'File {i}: {filename}')
```

```python
# continue: skip the rest of the loop body for this item and move to the next
# break: stop the loop entirely

rows = [
    {'order_id': '1001', 'quantity': '2',   'status': 'completed'},
    {'order_id': '1002', 'quantity': '-1',  'status': 'completed'},  # invalid qty
    {'order_id': '1003', 'quantity': '1',   'status': 'cancelled'},  # skip cancelled
    {'order_id': '1004', 'quantity': '3',   'status': 'completed'},
]

valid_rows = []
for row in rows:
    if row['status'] == 'cancelled':
        print(f'Order {row["order_id"]}: skipping cancelled order')
        continue   # go directly to the next row

    if int(row['quantity']) <= 0:
        print(f'Order {row["order_id"]}: skipping invalid quantity')
        continue

    valid_rows.append(row)

print(f'\nValid rows: {len(valid_rows)}')
```

```python
# Validate a list of rows and count problems
rows = [
    {'order_id': '1001', 'quantity': '2',  'unit_price': '1200.00'},
    {'order_id': '1002', 'quantity': '-1', 'unit_price': '35.00'},
    {'order_id': '1003', 'quantity': '0',  'unit_price': '75.00'},
    {'order_id': '1004', 'quantity': '1',  'unit_price': '400.00'},
]

problems = 0
for row in rows:
    qty = int(row['quantity'])
    if qty <= 0:
        print(f'Order {row["order_id"]}: invalid quantity ({qty})')
        problems += 1

print(f'\n{problems} problem rows found')
```

## 1.6 Functions

A function is a named, reusable block of code. In a data pipeline, every cleaning step becomes a function — you write the logic once and call it for every row across every file.

**Rule of thumb:** if you are writing the same code more than twice, put it in a function.

```python
def clean_price(value):
    '''Convert a price string to float. Returns None if the value cannot be parsed.'''
    try:
        cleaned = str(value).replace('$', '').replace(',', '').strip()
        return float(cleaned)
    except ValueError:
        return None  # return None instead of crashing

print(clean_price('$1,200.00'))  # 1200.0
print(clean_price('350.00'))     # 350.0
print(clean_price('N/A'))        # None — safe
print(clean_price(''))           # None — safe
```

```python
# Default parameter values: make parameters optional
def clean_text(value, case='title'):
    '''Strip whitespace and normalise the case of a string.'''
    cleaned = str(value).strip()
    if case == 'lower':
        return cleaned.lower()
    elif case == 'upper':
        return cleaned.upper()
    return cleaned.title()   # default

print(clean_text('  ELECTRONICS  '))             # 'Electronics'
print(clean_text('  electronics  ', 'lower'))    # 'electronics'
print(clean_text('  home & kitchen  '))          # 'Home & Kitchen'
```

```python
# Functions can call other functions
# clean_category uses both clean_text and the TYPO_MAP lookup

TYPO_MAP = {
    'Electrnics':    'Electronics',
    'Home & Kithen': 'Home & Kitchen',
}

def clean_category(value):
    '''Normalise a category: strip, title case, then fix known typos.'''
    normalised = clean_text(value)                        # calls clean_text
    return TYPO_MAP.get(normalised, normalised)           # fix typos if found

test_values = ['ELECTRONICS', '  Electrnics  ', 'clothing', 'Home & Kithen']
for v in test_values:
    print(f'  {v!r:25} -> {clean_category(v)}')
```

```python
# A function can return multiple values
def validate_row(row):
    '''Check a row for common data problems. Returns (is_valid, reason).'''
    try:
        qty = float(row.get('quantity', 0))
    except ValueError:
        return False, 'quantity is not a number'

    if qty <= 0:
        return False, f'quantity is {qty} (must be > 0)'

    return True, 'OK'

test_rows = [
    {'order_id': '1001', 'quantity': '2'},
    {'order_id': '1002', 'quantity': '-1'},
    {'order_id': '1003', 'quantity': 'abc'},
]

for row in test_rows:
    valid, reason = validate_row(row)
    status = 'OK' if valid else 'SKIP'
    print(f'Order {row["order_id"]}: [{status}] {reason}')
```

```python
# Putting it all together: apply functions to a list of rows

raw_rows = [
    {'order_id': '1001', 'category': '  ELECTRONICS  ', 'unit_price': '$1,200.00', 'quantity': '2'},
    {'order_id': '1002', 'category': 'clothing',         'unit_price': '35.00',     'quantity': '-1'},
    {'order_id': '1003', 'category': 'HOME & KITCHEN',   'unit_price': '$80.00',    'quantity': '1'},
    {'order_id': '1004', 'category': 'Electrnics',       'unit_price': '$150.00',   'quantity': '1'},
]

cleaned_rows = []
for row in raw_rows:
    row['category']   = clean_category(row['category'])   # uses clean_text inside
    row['unit_price'] = clean_price(row['unit_price'])

    valid, reason = validate_row(row)
    if not valid:
        print(f'Skipping order {row["order_id"]}: {reason}')
        continue

    cleaned_rows.append(row)

print(f'\nClean rows: {len(cleaned_rows)}')
for r in cleaned_rows:
    print(f'  {r}')
```

## Key Takeaways

1. **Variables** store values. Python figures out the type automatically.
2. **CSV files give everything as strings.** Always convert with `int()` or `float()` before doing maths. Never compare numeric strings with `>` or `<` — `'10' > '9'` gives the wrong answer.
3. **String methods** like `.strip()`, `.lower()`, `.title()`, `.replace()`, `.split()` are your main tools for cleaning text. Chain them.
4. **f-strings** (`f'Total: {total:,.2f}'`) are the clearest way to format output.
5. **Lists** hold ordered collections. Build filtered lists with `[x for x in items if condition]`. Sort with `sorted()`.
6. **Dictionaries** are lookup tables. Use `.get(key, default)` to safely handle missing keys. A typo correction map is a dict.
7. **`continue`** skips the rest of the loop body for that item. **`break`** stops the loop entirely.
8. **Functions** make code reusable. Functions can call other functions — `clean_category` calls `clean_text`.
9. **`try/except`** prevents one bad row from crashing the whole pipeline. Return `None` when a value cannot be fixed.
